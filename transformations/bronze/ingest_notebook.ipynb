```python
VOLUME_PATH = "/Volumes/marathos/bronze/raw"
```

```python
# reading in raw data from volume.

df = spark.read.csv(
    f"{VOLUME_PATH}/TWO_CENTURIES_OF_UM_RACES.csv",
    header=True,
    inferSchema=True
)
```

```python
# I have to rename the columns to be Delta compatible.

df_clean = df.toDF(
    "year_of_event", "event_dates", "event_name", "event_distance_length", "event_number_of_finishers", "athlete_performance", "athlete_club", "athlete_country", "athlete_year_of_birth", "athlete_gender", "athlete_age_category", "athlete_average_speed", "athlete_id"
)

# now im writing bronze delta table 
df_clean.write.format("delta").saveAsTable("marathos.bronze.races")
```
