```python
VOLUME_PATH = "/Volumes/marathos/bronze/raw"
```

```python
# Reading in the dataset from the marathon dataset

df = spark.read.csv(
    f"{VOLUME_PATH}/TWO_CENTURIES_OF_UM_RACES.csv",
    header=True,
    inferSchema=True
)
df.display()
```

```python
# Number of rows and columns 
print(f"Rows: {df.count()}" )
print(f"Columns: {len(df.columns)}")
```

```python
# Checking schema to see if it is a inferred
df.printSchema()
```

```python
# Descriptive summary of the numerical fields.
df.describe().display()
```

```python
# count on the nulls for each feild. 

from pyspark.sql.functions import col, sum as spark_sum

df.select([
    spark_sum(col(c).isNull().cast("int")).alias(c) 
    for c in df.columns
]).display()

# note for myself: for the futute the athleteclub column has nearly 3 million nulls.
```

```python
# How many unique events are there in the data?

df.select("Event name").distinct().count() # --> 260907 unique events.
```

```python
# How are ages distributed among  the runners
df.groupBy("Athlete age category").count().display() 
# i noticed that the silver layer has over 1 million null values.
```

```python
# Wich countries are most represented?
df.groupBy("Athlete country").count().orderBy("count", ascending=False).display()
```
